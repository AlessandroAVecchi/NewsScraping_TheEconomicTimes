```python
#importing libraries

import pandas as pd
import requests
import bs4
```

## The Economic Times - Finance

```python
#on The Economic Times there are different webpages of interest
```

```python
#retrieving the HTML of the webpage and parsing it

tet_finance = requests.get('https://economictimes.indiatimes.com/news/economy/finance').text
tet_finance
```

```python
soup_tet_finance = bs4.BeautifulSoup(tet_finance, 'html.parser')
soup_tet_finance
```

```python
#saving the tag and class of interest for the analysis 
#legend: art_tet_finance = articles, the economic times, finance

art_tet_finance = soup_tet_finance.find_all('div', {'class': 'eachStory'})
art_tet_finance
```

```python
#extracting and storing the data 

title_tet_finance = []
paragraph_tet_finance = []
date_tet_finance = []
url_tet_finance = []

for article in art_tet_finance:
    title_tet_finance.append(article.find_all('a')[1].text)
    paragraph_tet_finance.append(article.find('p').text)
    date_tet_finance.append(article.find('time', {'class':'date-format'}).text)
    url_tet_finance.append(article.find('a')['href'])
```

```python
url_tet_finance
```

```python
base_url = 'https://economictimes.indiatimes.com/news/economy/finance'
```

```python
for i in range(len(url_tet_finance)):
    url_tet_finance[i] = base_url + url_tet_finance[i]
```

```python
url_tet_finance
```

```python
title_tet_finance
```

```python
paragraph_tet_finance
```

```python
date_tet_finance
```

```python
#checking for lenght

print(len(title_tet_finance),
      len(paragraph_tet_finance), 
      len(date_tet_finance), sep="\n")
```
